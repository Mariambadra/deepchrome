# Preprocessing

## Build the Feature Matrix

```python
# Data Path
data_dir = '/kaggle/input/datasets/mariamali12/deepchrome-e047-counts/counts/'
```

```python
#The 5 Histone Markers
marks = ['H3K4me3', 'H3K4me1', 'H3K36me3', 'H3K9me3', 'H3K27me3']
# Load the 5 count files in one list
dfs = []
for mark in marks:
    count_file = f'{data_dir}/E047-{mark}_counts.txt'
    print(f"  Loading {mark}...")
    
    df = pd.read_csv(count_file,
                     sep='\t',
                     header=None,
                     names=['chr', 'start', 'end', 'bin_id', mark])
    dfs.append(df)
```

```python
dfs[0].head()
```

```python
# Merge all 5 dfs on bin_id
merged = dfs[0][['bin_id', marks[0]]].copy()
for i in range(1, len(marks)):
    merged = merged.merge(dfs[i][['bin_id', marks[i]]], on='bin_id')

print(f"Merged shape: {merged.shape}")
```

```python
# Extract gene ID from bin_id
merged['gene_id'] = merged['bin_id'].str.rsplit('_', n=1).str[0]
# Extract bin number from bin_id 
merged['bin_num'] = merged['bin_id'].str.rsplit('_', n=1).str[1].astype(int)

# Sort by gene_id and bin_num
merged = merged.sort_values(['gene_id', 'bin_num'])
```

```python
merged.head()
```

```python
# Build feature matrix
feature_rows = []
gene_ids = []

for gene_id, group in merged.groupby('gene_id', sort=False):
    group = group.sort_values('bin_num')
    features = group[marks].values.flatten()
    feature_rows.append(features)
    gene_ids.append(gene_id)
```

```python
# Build the final matrix
feature_matrix = pd.DataFrame(feature_rows, index=gene_ids)

print(f"\nFeature matrix shape: {feature_matrix.shape}")
```

```python
feature_matrix.head()
```

Each row is one gene, each column is a feature


```python
# checking the histone mark signal for the first 2 genes ENSG00000000003.10 and ENSG00000000005.5
import matplotlib.pyplot as plt

marks = ['H3K4me3', 'H3K4me1', 'H3K36me3', 'H3K9me3', 'H3K27me3']

fig, axes = plt.subplots(2, 1, figsize=(12, 8))

for idx in range(2):
    gene = feature_matrix.iloc[idx]
    gene_id = feature_matrix.index[idx]
    
    # Reshape to (100, 5)
    signal = gene.values.reshape(100, 5)
    
    for i, mark in enumerate(marks):
        axes[idx].plot(signal[:, i], label=mark)
    
    axes[idx].set_xlabel('Bin (1-100)')
    axes[idx].set_ylabel('Read Count')
    axes[idx].set_title(f'Histone Mark Signals — {gene_id}')
    axes[idx].legend()
    axes[idx].axvline(x=50, color='black', linestyle='--', 
                      alpha=0.3, label='TSS center')

plt.tight_layout()
plt.show()
```

Gene 1 — ENSG00000000003.10 (Top):
* H3K4me3 → Sharp peak at bins 10-20  ✅Strong H3K4me3 promoter signal
* H3K27me3 & H3K9me3 → Low and flat             ✅No repressive marks
Classic actively transcribed gene pattern. Likely HIGH expression (+1).

Gene 2 — ENSG00000000005.5 (Bottom):
* H3K4me3  → Completely flat, near zero  ❌
* H3K27me3 → Dominant signal, noisy      ⚠️
* H3K9me3  → Present throughout          ⚠️
* H3K4me1  → Very low                    ❌
Likely LOW expression (-1)

This is the pattern that our CNN model should learn from for the 20343 gene.

# Generate Expression Labels

```python
# Load RPKM file
# Correct way to load
rpkm_file = '/kaggle/input/datasets/mariamali12/deepchrome-e047-counts/57epigenomes.RPKM.pc/57epigenomes.RPKM.pc'
rpkm = pd.read_csv(rpkm_file, sep='\t', index_col=0)

print("Correct load:")
print(f"Shape: {rpkm.shape}")
print(f"First few columns: {rpkm.columns[:5].tolist()}")
print(f"Index name: {rpkm.index.name}")
print(f"\nENSG00000000003 E047 value: {rpkm.loc['ENSG00000000003', 'E047']}")
print(f"Median E047: {rpkm['E047'].median():.4f}")

# Regenerate labels correctly
e047_rpkm = rpkm['E047']
median = e047_rpkm.median()
labels = pd.Series(
    np.where(e047_rpkm >= median, 1, -1),
    index=e047_rpkm.index,
    name='label'
)

print(f"\nLabel for ENSG00000000003: {labels.loc['ENSG00000000003']}")
print(f"Label distribution: {labels.value_counts().to_dict()}")
```

```python
# Strip version numbers from feature matrix index
# ENSG00000186092.4 → ENSG00000186092
feature_matrix.index = feature_matrix.index.str.rsplit('.', n=1).str[0]

# Strip version numbers from labels index too
labels.index = labels.index.str.rsplit('.', n=1).str[0]

# Find common genes between both
common_genes = feature_matrix.index.intersection(labels.index)
print(f"Genes in feature matrix:  {len(feature_matrix)}")
print(f"Genes in RPKM file:       {len(labels)}")
print(f"Common genes:             {len(common_genes)}")
```

```python
# Filter both to common genes only
X = feature_matrix.loc[common_genes]
y = labels.loc[common_genes]

print(f"\nFinal X shape: {X.shape}")
print(f"Final y shape: {y.shape}")
print(f"\nLabel distribution in final dataset:")
print(y.value_counts())
```

```python
# Combine X and y into one dataframe
dataset = X.copy()
dataset['label'] = y

print(f"Final dataset shape: {dataset.shape}")
print(f"Label column check: {dataset['label'].value_counts().to_dict()}")
```

```python
# Save to Kaggle working directory
output_path = '/kaggle/working/E047_dataset.csv'
dataset.to_csv(output_path)
```
